```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
plt.plot([1,2,3,4,5])
```

```python
t = np.linspace(0, 10, 100)  # This creates an array of 100 values from 0 to 10
plt.plot(t, t**2)
plt.show()  # Display the plot
```

```python
a = [1,2,3,4,5]
b = [1,4,9,16,25]
plt.plot(a,b)
```

```python
a = [1,2,3,4,5]
b = map(lambda x: x**2, a)
# print(list(b))
plt.plot(a,list(b))
```

```python
a = np.linspace(-2* math.pi, 2 * math.pi, 100)
b = map(lambda x: math.sin(x), a)
# print(list(b))
plt.plot(a,list(b))
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
fig, ax=plt.subplots()
ax.plot(x,y,'r')
ax.set_title('A single plot')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.histplot(data=titanic, x='parch', shrink=0.5, hue='sex', palette=['yellow', 'blue'], discrete=True)
```

```python
sns.set_style('whitegrid')
sns.histplot(titanic['survived'], discrete=True)
```

```python
# create a graph that helps understand how both sex and passenger class affected survival on the titanic
sns.barplot(x='sex', y='survived', hue='pclass', data=titanic)
```

```python
sns.boxplot(x='survived', y='age', data=titanic)
```

```python
sns.pairplot(titanic, hue='sex', diag_kind='hist', kind='scatter', palette='husl')
```
