```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
sunshine = np.array([2,3,5,7,9])
ice_cream = np.array([4,5,7,10,15])

slope, intercept = np.polyfit(sunshine, ice_cream, 1)
print(f'slope: {slope}, intercept: {intercept}')

plt.scatter(sunshine, ice_cream)
plt.plot(sunshine, slope*sunshine + intercept, color='red')
```

```python
# do the samething with seaborn
sns.regplot(x=sunshine, y=ice_cream)
```

```python
# use a list comprehension to plot a sine wave...between 0 and 2PI
# x = np.linspace(0, 2*math.pi, 100)
# y = np.sin(x)
# plt.plot(x, y)

x = np.linspace(0, 2*math.pi, 100)
y = [math.sin(i) for i in x]
sns.lineplot(x=x, y=y, color='red', label='sin(x)')
```

```python
titanic = sns.load_dataset('titanic')
g = sns.FacetGrid(titanic, col='survived', row='pclass')
g.map(plt.hist, 'age')
```

```python
sns.jointplot(x='age', y='fare', data=titanic)
```

```python
# Select only the numerical columns
numerical_cols = titanic.select_dtypes(include=[np.number])

# Calculate the correlation matrix
tc = numerical_cols.corr()

# Create the heatmap
sns.heatmap(tc, cmap='coolwarm')
```

```python
sns.stripplot(x='sex', y='age', hue='survived', data=titanic)
```
