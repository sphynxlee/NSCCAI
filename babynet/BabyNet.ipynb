```python
# Fahrenheit to Celsius "predictor"

import random
# slope
weight1 = random.randint(1, 10)
print('weight1 is: ', weight1)

# y-intercept(b)
weight2 = -18
learning_rate = 0.01

data = [[87, 31], [97, 37], [22, -5], [100, 39], [32, 0]]

def baby_net(input):
  # return m*x + b
  output = weight1 * input + weight2
  return output

# cost function, energy function, error function, loss function, everything under the sun function
def error(pred, act):
  err = act - pred
  return err

# # experimenting/testing our error function using some sample data
# sample_input = data[0][0]
# sample_label = data[0][1]
# one_error = error(baby_net(sample_input), sample_label)
# # print(baby_net(100))
# print("Error: ", one_error)
# print("Fahrenheit input: ", sample_input)
# print("Predicted celsius equivalent: ", baby_net(sample_input))

def backpropagation(err, input):
  # change in slope/weight1
  delta_m = learning_rate * err/input
  return delta_m

# training loop -where we look at all the data, pass it into baby_net
# calculate the errors, backpropagate accordingly, and do it over and over again
# hoping that baby_net will get smarter
# NUMBER_OF_ITERATIONS = 1000
NUMBER_OF_EPOCHS = 100
for i in range(NUMBER_OF_EPOCHS):
  print("\n**************** Epoch ", i, " *****************")
  for input, label in data:
    print("\t input: ", input, "label: ", label)
    prediction = baby_net(input)
    print("\t Prediction: ", prediction)
    err = error(prediction, label)
    print("\t Error: ", err)
    # adjust the mode weights, we have only one - the slope
    # we do this by updating the global weight1
    delta_m = backpropagation(err, input)
    weight1 = weight1 + delta_m
    print("\t Weight1(slope): ", weight1)
    print("\n")


```
